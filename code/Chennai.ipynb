```python
from pycaret.regression import *
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/user/Downloads/weather_data/Chennai.csv')
reg = setup (data=data, target='T2M', session_id=123)
```

```python
best=compare_models(exclude=['ransac'],sort='MSE')
```

```python
create_xg= create_model('xgboost')
```

```python
tuned_xg = tune_model(create_xg)
```

```python
plot_model(tuned_xg)
plot_model(tuned_xg, plot = 'error')
plot_model(tuned_xg, plot = 'feature')
```

```python
evaluate_model(tuned_xg)
```

```python
predict_model(tuned_xg)
```

```python
final_xg = finalize_model(tuned_xg)
print(final_xg)
predict_model(final_xg)
```

```python
unseen_predictions = predict_model(final_xg, data=data)
unseen_predictions.head()
```

```python
save_model(final_xg,'Final XGBoost Model 18Sep2023')
```
